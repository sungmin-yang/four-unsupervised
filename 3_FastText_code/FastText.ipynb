```python
# written in ------- Python3 ----------
# In order to run this code, need Cython, fastText library.
# Executed in Linux/MacOS
# Follow this website instuction : https://github.com/facebookresearch/fastText/tree/master/python
# $ pip3 install Cython
# $ git clone https://github.com/facebookresearch/fastText.git
# $ cd fastText
# $ pip install .

from __future__ import absolute_import
from __future__ import division
from __future__ import print_function
from __future__ import unicode_literals
from __future__ import division, absolute_import, print_function

import os
import fastText
from fastText import train_supervised
import numpy as np
import datetime, os
import numpy as np
from sklearn import metrics
from sklearn.metrics import accuracy_score
    
RANDOM_SEED = 2212
np.random.seed(RANDOM_SEED)
```

```python
# ---------------- functions for FastText ---------------
def print_results(N, p, r):
    print("N\t" + str(N))
    print("P@{}\t{:.3f}".format(1, p))
    print("R@{}\t{:.3f}".format(1, r))
    
    
def fastText_pred(test_file_path, trained_model):
    y_true = []
    y_pred = []
    test_texts = []

    with open(test_file_path, 'r', encoding='utf8') as f:
        for line in f:
            words = line.split()
            y_true.append(words[0])
            test_texts.append(' '.join(words[1:]))

    for sentence in test_texts:
        y_pred.append(trained_model.predict(sentence)[0][0])
    return y_true, y_pred
# ---------------- End of functions ---------------

    
    
# ---------------- Hyperparameters setting ----------------
instance_list = [10, 50, 100, 200, 500, 1000, 2000, ""]
target_list = ['agnews', 'dbpedia', 'amazon_six', 'tripadvisor', 'yahoo_a', 'yelp_f']
lr_list     = [1] # It can handle multiple values e.g., [0.1, 0.25, 0.5, 1]
word_ngram_list = [2]
minCount_list  = [2]
epoch=15
verbose=2


# ---------------- Training  ----------------
for target in target_list:
    
    train_file_path_list = []
    for instance in instance_list:
        train_file_path = "./data/data_fasttexified/"+target+"_train{}_fastText.txt".format(instance)
        train_file_path_list.append(train_file_path)
        pretrained_model = "./data/data_fasttexified/"+target+"_train{}_fastText.txt".format(instance)
    test_file_path  = "./data/data_fasttexified/"+target+"_test{}_fastText.txt".format(instance)

    for lr in lr_list:
        for wordNgrams in word_ngram_list:
            for minCount in  minCount_list:
                for train_file_path, instance in zip(train_file_path_list, instance_list):
                    title = target+'.report'

                    #  load previous vectors
                    # pretrained_vectors_by_Facebook = './fastText/data/fastText_wiki-news-300d-1M.vec'
                    pretrained_vectors = '{}.model.vec'.format(target)

                    model = fastText.train_supervised(input=train_file_path, epoch=epoch, lr=lr, wordNgrams=wordNgrams, verbose=verbose, minCount=minCount,
                                                           pretrainedVectors=pretrained_vectors)

                    y_true, y_pred = fastText_pred(test_file_path, model)


                    # description = "Quick accurcy : {:3f}%\n\n".format(np.mean(y_true == y_pred)*100)
                    description = "\n instance per class {}, epoch: {}, learning rate : {}, word-ngrams : {} minCount: {}\n".format(instance if instance != "" else "total", epoch, lr, wordNgrams, minCount)

                    # description += "Sample of Y_test : {}, \n Y_pred : {}\n".format(y_true[:10], y_pred[:10])

                    print('\n\n\nDescription : '+ str(description) + '\n')
                    print(txt2 = "*"*30 + title +"*"*30 + '\n')
                    print(txt3 = str(metrics.classification_report(y_true, y_pred)) + '\n')
                    print(txt4 = "Accuracy : "+ str(accuracy_score(y_true, y_pred)))
    
```
